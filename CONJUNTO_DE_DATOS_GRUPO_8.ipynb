# Obtención de datos
**Proyecto:** Turistificación y transformación del comercio local en Medellín

Este notebook carga las tres bases de datos originales directamente desde el repositorio de GitHub.

---

**Bases de datos:**
- `BD1` — Estructura empresarial de Medellín según comunas y actividad económica *(Cámara de Comercio de Medellín / datos.gov.co)*
- `BD2` — Extranjeros no residentes que ingresan a Colombia *(MinCIT)*
- `BD3` — Llegada de pasajeros en vuelos nacionales *(MinCIT)*

---
**Nota aclaratoria:** La base de datos encontrada en el sitio web del Ministerio de comercio, industria y turismo no es la original pero los datos sí. Tuvimos que separar los datos en archivos distintos ya que el documento encontrado contiene mucha información que no es relevante para nuestro estudio.

## Carga de bases de datos originales desde GitHub

Los archivos están almacenados en el repositorio del proyecto llamado "Turismo-Medellin-datos". Se leen directamente desde la URL pública de GitHub usando pandas.

```
# Libreía Pandas
import pandas as pd

# Cargamos Urls
BD1 = "https://raw.githubusercontent.com/Ana-C-Zapata/Turismo-Medellin-datos/main/BD1_Estructura_Empresarial_Medellin_Actividad_Economica.csv"
BD2 = "https://raw.githubusercontent.com/Ana-C-Zapata/Turismo-Medellin-datos/main/BD2_Extranjeros_no_residentes_que_ingresan_a_Colombia.csv"
BD3 = "https://raw.githubusercontent.com/Ana-C-Zapata/Turismo-Medellin-datos/main/BD3_Llegada_de_pasajeros_en_vuelos_nacionales.csv"
```

## Carga de Datasets

```
# Cargar bases de datos desde GitHub
# Usamos encoding='latin-1' para evitar errores con tildes y caracteres especiales
df1 = pd.read_csv(BD1, encoding='latin-1')
df2 = pd.read_csv(BD2, encoding='latin-1')
df3 = pd.read_csv(BD3, encoding='latin-1')
```
